# Classic Alarm Clock Project

```python
import time
from datetime import datetime
import pygame

alarm_input = input("Enter the alarm time ( HH:SS AM/PM ) : ").strip().upper()
print(f"Alarm set for : {alarm_input}")

pygame.mixer.init()

while True :

    now = datetime.now()

    current_time = now.strftime("%I:%M %p")
    print(f"Current Time : {current_time}")

    print(f"Checking {current_time} vs {alarm_input}")

    if current_time == alarm_input :
        print("Wake Up")
        try :
            pygame.mixer.music.load("air_raid_siren.mp3")
            pygame.mixer.music.play(loops=-1)
            input("Press Enter to stop the alarm")
            pygame.mixer.music.stop()
            print("Alarm stopped")
        except Exception as e :
            print(f"Alarm did not work due to error : {e}")
        break
    time.sleep(1)
```
